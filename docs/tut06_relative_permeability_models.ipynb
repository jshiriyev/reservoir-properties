# Three-phase relative permeability models.

A three-phase water-oil-gas system is described by the following empirical parameters obtained from oil-water and oil-gas two-phase experiments:

$S_\text{wr}=0.1$,
$S_\text{orw}=0.4$,
$S_\text{org}=0.2$,
$S_\text{gr}=0.05$,
$k^0_\text{rw}=0.3$,
$k^0_\text{row}=0.8$,
$k^0_\text{rog}=0.8$,
$k^0_\text{rg}=0.3$,
$N_\text{w}=2.0$,
$N_\text{g}=2.0$,
$N_\text{ow}=2.0$,
$N_\text{og}=2.0$

Compute three-phase relative permeabilities ($k_\text{rw}$, $k_\text{ro}$, $k_\text{rg}$) at $S_{w}=0.3$, $S_{o}=0.5$, $S_{g}=0.2$.

Reference: Balhoff's book Example 1.2 

```python
rp = rperm.StonesI(swr=0.1,sor_ow=0.4,sor_go=0.2,sgr=0.05,k0rw=0.3,k0ro_ow=0.8,k0ro_go=0.8,k0rg=0.3)
```

```python
rp.get(0.3,0.5,0.2)
```

## Water Relative Permeability

```python
def kr_w(saturations):
    krw,_,_ = rp.get(*saturations) 
    return krw
```

```python
# Plot ternary diagram
fig, ax = plt.subplots(figsize=(7, 6))

scale = 20

tax = ternary.TernaryAxesSubplot(ax=ax, scale=scale)

# Scatter plot with color mapping
tax.heatmapf(kr_w, cmap="cividis", colorbar=True)

# Set labels
ax.text(x=scale,y=0,s="$S_w$",ha='left',va='top',fontsize=12)
ax.text(x=scale/2,y=scale*np.sin(np.deg2rad(60)),s="$S_o$",ha='center',va='bottom',fontsize=12)
ax.text(x=0,y=0,s="$S_g$",ha='right',va='top',fontsize=12)

# Plot settings
tax.boundary(linewidth=1.0)
tax.gridlines(color="black", multiple=1)

tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')

plt.show()
```

## Oil Relative Permeability

```python
def kr_o(saturations):
    _,kro,_ = rp.get(*saturations) 
    return kro
```
