# Oleic Phase Properties

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from respy import OilPhase
```

```python
p = np.linspace(0,3000,1000)
```

```python
bpp = 1666.
```

```python
sgsg = 22/28.96
gAPI = OilPhase.sgco_to_gAPI(0.822)
```

```python
po = OilPhase(sgsg=sgsg, gAPI=gAPI, temp=150.)
```

```python
Rs = po.gass(p,bpp)
Bo = po.fvf(p,bpp)
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlim((0,3000))

ax1.plot(p,Bo,'k')
ax1.set_ylim((1,1.4))

ax2 = ax1.twinx()

ax2.plot(p,Rs,'k--')
ax2.set_ylim((0,500))

plt.show()
```

```python
co = po.comp(p,bpp)
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlim((0,3000))

ax1.plot(p,co,'k')
ax1.set_ylim((0,5e-4))

# ax2 = ax1.twinx()

# ax2.plot(p,Rs,'k--')
# ax2.set_ylim((0,500))

plt.show()
```
